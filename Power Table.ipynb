```python


cont = True
print("Learn your squares and cubes!")

while cont:
    num = int(input("Please enter an integer:"))
    underline = "======="
    print('Number\t\tSquared\t\tCubed')
    print('=======\t\t=======\t\t=======')

    for i in range(num):
        k= i+1
        print(f'{k}\t\t{k*k}\t\t{k*k*k}')
    
    
    printstr = "\t"
    # creating multiplaction table
    for i in range(num):
        printstr += f'{i+1}\t'
    print(printstr)
    printstr= "\t"
    for i in range(num):
        printstr += f'=\t'
    print(printstr)
    printstr= " "
    for i in range(num):
        printstr += f'{i+1} |\t'
        for j in range(num):
            printstr += f'{(i+1)*(j+1)}\t'
        print(printstr)
        printstr= " "
    print(printstr)
    
    con = input("continue?(y/n)")
    if con == "n":
        break
    
    
```
